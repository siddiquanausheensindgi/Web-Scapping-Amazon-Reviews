```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
link = 'https://www.amazon.in/OnePlus-Mirror-Black-128GB-Storage/product-reviews/B07DJHV6VZ/ref=cm_cr_arp_d_viewpnt_rgt?ie=UTF8&reviewerType=all_reviews&filterByStar=critical&pageNumber=1'
```

```python
page = requests.get(link)   #loading the page
```

```python
page.content  # the out obtained 
```

```python
soup =  bs(page.content, 'html.parser')
```

```python
print(soup.prettify())
```

# Scrapping the name of the user

```python
names = soup.find_all('span', class_='a-profile-name')
```

```python
names
```

```python
cust_name = []                                     #to get the customers names only
for i in range (0, len(names)):
    cust_name.append(names[i].get_text())
```

# Scrapping the Review title

```python
#title = soup.find_all('span', class_='review-title')      #span is the class name
#title                            
```

```python
title = soup.find_all('a', class_='review-title-content')      # review title.. a is the parent class 
title                                   # now here we get the review
```

```python
review_title = []                           
for i in range (0, len(title)):
    review_title.append(title[i].get_text())#now save the data
review_title
```

```python
review_title[:] = [titles.lstrip('\n') for titles in review_title]   #removing \n from LHS side
#review_title[:] = [titles.lstrip('\n') for titles in review_title]
review_title
```

```python
review_title[:] = [titles.rstrip('\n') for titles in review_title] #removing \n from RHS side
review_title
```

```python
len(review_title)
```

# Scrapping the Rating

```python
######### Rating

rating = soup.find_all('i', class_='review-rating') # to know review rating 
rating
                      
```

```python
rate = []                           
for i in range (0, len(rating)):
    rate.append(rating[i].get_text())#now save the data
rate
```

# Scrapping the Review content

```python
review = soup.find_all("span", {"data-hook":"review-body"})
review                           #Review body conent ...  using the dictionary data-hook from the inspect of
```

```python
#using the loop and extracting the actual content
review_content = []                           
for i in range (0, len(review)):
    review_content.append(review[i].get_text())#now save the data
review_content
```

```python
# taking out the \n in the above chunk output content from the left hand side

review_content[:] = [reviews.lstrip('\n') for reviews in review_content]
review_content
```

```python
# taking out the \n from the output content from the right hand side


review_content[:] = [reviews.rstrip('\n') for reviews in review_content]
review_content
```

```python
len(review_content)
```
